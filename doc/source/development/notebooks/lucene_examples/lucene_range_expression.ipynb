```python
from logprep.filter.lucene_filter import LuceneFilter
```

```python
def print_match(expression, value):
    lucene_filter = LuceneFilter.create(expression)
    print("Range:", lucene_filter)

    matching = lucene_filter.matches({"key": value})
    print(f"Value: {value} ({'match' if matching else 'no match'})", end="\n\n")
```

Example: lucene **`int`** range expression:

```python
lucene_range_expression = "key:[5 TO 10]"
# match
print_match(expression=lucene_range_expression, value=10)

# no match:
print_match(expression=lucene_range_expression, value=11)
```

Example: lucene **`float`** range expression:

```python
lucene_range_expression = "key:[3.7 TO 5.5]"
# match
print_match(expression=lucene_range_expression, value=3.8)

# no match:
print_match(expression=lucene_range_expression, value=5.6)
```

Example: lucene **`str`** range expression:

```python
lucene_range_expression = "key:[A TO K]"
# match
print_match(expression=lucene_range_expression, value="Bernhard")

# no match:
print_match(expression=lucene_range_expression, value="Leo")
```

Example: lucene **`ISO 8601 timestamp`** range expression:

```python
lucene_range_expression = "key:[2026-01-01 TO 2026-01-31]"
# match
print_match(expression=lucene_range_expression, value="2026-01-01")

# no match:
print_match(expression=lucene_range_expression, value="2026-02-01")

lucene_range_expression = "key:[2026-01-01T06:00Z TO 2026-01-01T18:00Z]" # one day
# match
print_match(expression=lucene_range_expression, value="2026-01-01T18:00Z")

# no match:
print_match(expression=lucene_range_expression, value="2026-01-01T18:01Z")
```

Example: More complex expressions with `AND/OR`:

```python
# AND examples
lucene_range_expression = "key:[0 TO 10] AND key:[10 TO 15]"

# match
print_match(expression=lucene_range_expression, value=10)

# no match
print_match(expression=lucene_range_expression, value=11)

lucene_range_expression = "key:[0 TO 10} AND key:[10 TO 15]"
# no match
print_match(expression=lucene_range_expression, value=10)
```

```python
# OR examples
lucene_range_expression = "key:[0 TO 5] OR key:[10 TO 50]"

# match
print_match(expression=lucene_range_expression, value=2)
print_match(expression=lucene_range_expression, value=22)

# no match
print_match(expression=lucene_range_expression, value=7)
print_match(expression=lucene_range_expression, value=77)
```
